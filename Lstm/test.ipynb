```python
import yfinance as yf
data = yf.Ticker("^GSPC").history(period='1mo')
```

```python
import matplotlib.pyplot as plt

def kbar(open, close, high, low, pos):
    if close > open:             # 上漲
        color='red'                 # 紅 K 棒
        height=close - open   # 高度=收盤-開盤
        bottom=open             # 底部=開盤
    else:                               # 下跌
        color='green'              # 綠 k 棒
        height=open - close   # 高度=開盤-收盤
        bottom=close             # 底部=收盤
    # 繪製 k 棒實體      
    plt.bar(pos, height=height,bottom=bottom, width=0.5, color=color)
    # 繪製 k 棒上下影線
    plt.vlines(pos, high, low, color=color)

for i in data.index:
    kbar(data['Open'].loc[i], data['Close'].loc[i], data['High'].loc[i], data['Low'].loc[i], i)
# plt.arrow(data.index[0], data['Open'].iloc[0], 10, 100, width=0.5, color='red')
plt.annotate('', xy=(data.index[10], data['Open'].iloc[10]), xytext=(data.index[0], data['Open'].iloc[0]),
            arrowprops=dict(facecolor='red', arrowstyle='simple'))
plt.show()  
```
