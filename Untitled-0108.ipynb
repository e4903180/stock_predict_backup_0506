```python
#定義自動sarimax的函數
    #2層for迴圈帶入order跟seasonal_order，並將oder、seasonal_order與aic存入aic_list中
    #遍歷aic_list找出min(aic)與其超參數，最後顯現結果
def auto_sarimax(train_df, orders, seasonal_orders):
    aic_list=[]
    for order in orders:
        for seasonal_order in seasonal_orders:
            model= SARIMAX(train_df,
                            order=order,
                            seasonal_order=seasonal_order,
                            enforce_invertibility=False, enforce_stationarity=False)\
                    .fit(full_output=False, disp=False)

            aic_list.append([order,seasonal_order,model.aic])
            print(f'Current Status: order #{orders.index(order)+1}, \
                    seasonal_order #{seasonal_orders.index(seasonal_order)+1}, \
                    No.={(orders.index(order)+1)*(seasonal_orders.index(seasonal_order)+1)}')
    min_index=0
    for i in range(len(aic_list)):
        if aic_list[i][2]<=aic_list[min_index][2]:
            min_aic=i
    
    print(f'The Best Model: SARIMAX{aic_list[min_aic][0]}{aic_list[min_aic][1]} and AIC={aic_list[min_aic][2]}')
```
