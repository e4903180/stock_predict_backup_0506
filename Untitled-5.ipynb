```python
nums = [1,2,3,2,4,5,4,3,2,1,8,9,10,11,10,9,8]
peaks = []
troughs = []
for idx in range(1, len(nums)-1):
    if nums[idx-1] < nums[idx] > nums[idx+1]:
        peaks.append((idx, nums[idx]))
    if nums[idx-1] > nums[idx] < nums[idx+1]:
        troughs.append((idx, nums[idx]))
        
print(peaks) # [(2, 3), (5, 5), (13, 11)]
print(troughs) # [(3, 2), (9, 1)]
```

```python
nums = [1,2,3,10,4,5,4,3,2,1,8,9,10,11,10,9,8]
df = pd.Series(nums)
peaks = pd.Series()
troughs = pd.Series()
for idx in range(1, len(nums)-5):
    if nums[idx-5] < nums[idx] > nums[idx+5]:
        peaks.at[idx] = nums[idx]
    if nums[idx-5] > nums[idx] < nums[idx+5]:
        troughs.at[idx] = nums[idx]

i = 4
for idx in range(len(nums)-5, len(nums)):
    if nums[idx-5] < nums[idx] > nums[idx+i]:
        peaks.at[idx] = nums[idx]
    if nums[idx-5] > nums[idx] < nums[idx+i]:
        troughs.at[idx] = nums[idx]
    i = i - 1
        
print(peaks) # [(2, 3), (5, 5), (13, 11)]
print(troughs) # [(3, 2), (9, 1)]
```
