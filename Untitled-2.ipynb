```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack
```

```python
df_original = yf.download('^GSPC' ,period="100d" ,progress=False)
display(df_original)
df = df_original[['Close']]
# df = df.drop(df.tail(100).index)
display(df)
plt.plot(df)
```

```python

F = fftpack.fft(df)
N = df.size
freq = fftpack.fftfreq(N,d=1)
display(freq)
```

```python
mask = np.where(freq >= 0) #因波形對稱性，只需看半邊
fig, axes = plt.subplots(2, 1, figsize=(8, 6))
axes[0].plot(df,label="real")
axes[0].set_ylabel("Price ($)")
axes[0].set_xlabel("frequency (Day)")
axes[1].plot(freq[mask], abs(F[mask]), label="real")
axes[1].set_ylabel("$(|F|)$")
axes[1].set_xlabel("frequency")
plt.tight_layout()
plt.show()
```
