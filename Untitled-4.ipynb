```python
import numpy as np
import pylab as pl
from numpy import fft
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
date_0 = '2008-08-01'
date_1 = '2022-09-01'
date_2 = '2022-10-01'
stock = '0050.TW'

# date_0 = input('請輸入開始日期 : eg.2000-01-01')
# date_1 = input('請輸入結束日期 : eg.2022-02-01')
# date_2 = input('請輸入結束日期 : eg.2022-10-01')
# stock = input('請輸入股票代碼 : eg.0050.TW')

data = yf.Ticker(stock).history(start=date_0, end=date_1)
price = data['Close']
x = np.array(price)


data2 = yf.Ticker("0050.TW").history(start=date_0, end=date_2)
price2 = data2['Close']
x2 = np.array(price2)
```

```python
n_predict = price2.size - price.size
print(n_predict)
n = x.size
print(n)
n_harm = 100                     # number of harmonics in model
t = np.arange(0, n)
p = np.polyfit(t, x, 1)         # find linear trend in x
# x_notrend = x       # detrended x
x_notrend = x - p[0] * t        # detrended x
```

```python
x_freqdom = fft.fft(x_notrend, n)  # detrended x in frequency domain

pl.plot(x_notrend)
pl.show()
pl.plot(1/x_freqdom)
pl.show()
pl.plot(x_freqdom)
pl.show()
f = fft.fftfreq(n)
pl.plot(f)
pl.show()
```

```python
indexes = list(range(n))             # frequencies

# indexes.sort(key = lambda i: np.absolute(f[i]))  # sort indexes by frequency, lower -> higher
# sort indexes by amplitudes, lower -> higher
indexes.sort(key=lambda i: np.absolute(x_freqdom[i]))
indexes.reverse()  # sort indexes by amplitudes, higher -> lower

t = np.arange(0, n + n_predict)
restored_sig = np.zeros(t.size)
```

```python
mask = np.where(f >= 0)
print(mask)
pl.plt.figure(figsize=(15, 4))
# pl.xlim(0,0.01)
pl.plot(f[mask], abs(x_freqdom[mask]), label="real")
pl.legend()
pl.show()
pl.plt.figure(figsize=(15, 4))
# pl.xlim(0,0.01)
pl.plot(1/f[mask], abs(x_freqdom[mask]), label="real")
pl.legend()
pl.show()
# 做以DAY為X軸的頻譜圖
```

```python
for i in indexes[:1 + n_harm * 2]:
    ampli = np.absolute(x_freqdom[i]) / n   # amplitude
    phase = np.angle(x_freqdom[i])          # phase
    signal = ampli * np.cos(2 * np.pi * f[i] * t + phase)
    restored_sig += signal

    if (i < 5):
        print('i = ', i)
        print('fequence : ', abs(f[i]))
        print('Period : ', 1/abs(f[i]), '(Days)')
        fig, axes = plt.subplots(2, 1, figsize=(8, 6))
        axes[0].plot(signal, 'b', label='sig')
        axes[1].plot(restored_sig, 'r', label='restored_sig')
        plt.show()

# for i in indexes[:1 + n_harm * 2]:
#     print('i = ',i)
#     # print('x_freqdom = ',x_freqdom[i])
#     print('fequence : ' ,abs(f[i]))
#     print('Period : ' ,1/abs(f[i]) ,'(Days)')
#     ampli = np.absolute(x_freqdom[i]) / n   # amplitude
#     print('ampli = ' ,ampli)
#     phase = np.angle(x_freqdom[i])          # phase
#     print('phase = ' , phase)
#     # print('angle = ' , 2 * np.pi * f[i] * t + phase)
#     restored_sig += ampli * np.cos(2 * np.pi * f[i] * t + phase) #cos波為偶函數，若用sin波會相消
#     # print(np.cos(2 * np.pi * f[i] * t + phase))
#     # print('resig = ' ,restored_sig)
#     pl.plt.figure(figsize=(15,4))
#     pl.plot((ampli * np.cos(2 * np.pi * f[i] * t + phase)),label = 'harm')
#     pl.legend()
#     pl.show()
#     pl.plt.figure(figsize=(15,4))
#     pl.plot(price2.index, restored_sig, 'r',label = 'Predict')
#     pl.plt.grid(axis='x',linestyle='-.')
#     pl.legend()
#     pl.show()
# signal = restored_sig
signal = restored_sig + p[0] * t

# print('signal = ' ,signal)
# pl.plt.figure(figsize=(15,4))
# pl.plot(price.index, x, 'c', linewidth = 2, label = 'Train')
# pl.plot(price2.index, signal, 'r',label = 'Predict')
# pl.plt.grid(axis='x',linestyle='-.')
# pl.legend()
# pl.show()
```
